```python
#import libraries
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sbn
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
# Read in the data 

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
# Save file locally
urlretrieve(url, "wine.data")
df = pd.read_csv('wine.data', sep=',', header=None)
print(df.shape)
df.head()
```

First we need some more informative column names. 

The first column is the wine class and the other 13 columns are as follows: 

* Alcohol
* Malic acid
* Ash
* Alcalinity of ash  
* Magnesium
* Total phenols
* Flavanoids
* Nonflavanoid phenols
* Proanthocyanins
* Color intensity
* Hue
* OD280/OD315 of diluted wines
* Proline  

```python
df.columns = ["true_class", "alcohol", "acid", "ash", "alc_ash", "mag", "tot_phenol", 
              "flav", "nonf_phenol", "proanth", "color", "hue", "od", "proline"]
```
